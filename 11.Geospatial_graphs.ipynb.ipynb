**GEOCHART**

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load a shapefile (example: world countries)
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Plot the geochart
world.plot()
plt.title('Geochart with Geopandas')
plt.show()
```

**GEOGRAPHIC HEATMAP**

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Sample Data: You can replace this with your geospatial data
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Add a sample variable for heatmap intensity
world['heatmap_intensity'] = range(len(world))

# Plot the Geographic Heatmap
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world.plot(column='heatmap_intensity', cmap='YlOrRd', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)

# Add title and customize
plt.title('Geographic Heatmap Example', fontdict={'fontsize': '15', 'fontweight': '3'})

# Show the plot
plt.show()
```
